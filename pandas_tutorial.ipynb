```python
# Create a DataFrame
data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']
        }

df = pd.DataFrame(data)
#print(df)

# 1. Column Selection
print(df[['Name', 'Qualification']])
print(df.loc[:,['Name','Qualification']])
print(df.iloc[:,[0,1,2,3,4]]) # all rows, all columns

# 2. Row Selection
# Use loc[]
print("First row data with loc[0]")
print("-------------------------------")
print(df.loc[0])
print("First row data each column alone")
print("-------------------------------")
print("Name : ",df.loc[0,"Name"])
print("Age : ",df.loc[0,"Age"])
print("City : ",df.loc[0,"City"])
print("Address : ",df.loc[0,"Address"])
print("Qualification : ",df.loc[0,"Qualification"])
print(df.loc[0:2]) # rows from 0 to 2

# Use iloc[]
print("First row data with iloc[0]")
print("-------------------------------")
print(df.iloc[0])
print("First row data each column alone")
print("-------------------------------")
print("Name : ",df.iloc[0,0]) # first row, first column
print("Age : ",df.iloc[0,1])
print("City : ",df.iloc[0,2])
print("Address : ",df.iloc[0,3])
print("Qualification : ",df.iloc[0,4])
print("-------------------------------")
print(df.iloc[0,[0,1,2,3,4]]) #
print("-------------------------------")
print(df.iloc[0:2]) # rows from 0 to 2
```

```python
# Working with Missing Data
# 1. Checking for Missing Values using isnull() and notnull()
import pandas as pd

import numpy as np

dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]
        }

df = pd.DataFrame(dict)

df.isnull()

# 2. Filling Missing Values using fillna(), replace() and interpolate() or delete using dropna()
df.fillna(0)
```

## Pandas Aggregation Functions
Pandas DataFrames have several built-in functions for aggregation, which allow you to summarize your data. Here are some common ones:

*   `.sum()`: Calculates the sum of values in a column or row.
*   `.mean()`: Calculates the average of values.
*   `.median()`: Calculates the median of values.
*   `.min()`: Finds the minimum value.
*   `.max()`: Finds the maximum value.
*   `.count()`: Counts the number of non-null values.
*   `.std()`: Calculates the standard deviation.
*   `.var()`: Calculates the variance.
*   `.agg()` or `.aggregate()`: Allows you to apply multiple aggregation functions at once.

You can apply these functions to an entire DataFrame, a single column (Series), or grouped data.

```python
# Do a complete example please
# create dataframe an apply all these above functions

data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']
        }

df = pd.DataFrame(data)
display(df)

# apply these previous agg functions please

print("\nSum of Age:", df['Age'].sum())
print("Mean of Age:", df['Age'].mean())
print("Median of Age:", df['Age'].median())
print("Minimum Age:", df['Age'].min())
print("Maximum Age:", df['Age'].max())
print("Count of Age:", df['Age'].count())
print("Standard Deviation of Age:", df['Age'].std())
print("Variance of Age:", df['Age'].var())

# Using agg() for multiple aggregations
print("\nMultiple aggregations on Age:")
display(df['Age'].agg(['sum', 'mean', 'min', 'max']))

# value_count
print("\nValue counts of Age:")
display(df['Age'].value_counts())
```
